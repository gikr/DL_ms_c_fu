```python
%matplotlib inline
import matplotlib.pyplot as plt
import matplotlib as mpl
import numpy as np
import tae 
import pyemma

with np.load('dimredux-challenge-01-data.npz') as fh:
    data_x = fh['data_x']
    validation_x = fh['validation_x']
    validation_y = fh['validation_y']
    
    
```

```python
embedding, training_loss, validation_loss = tae.ae(
    data_x, dim=1, lag=1, n_epochs=300, validation_split=0.5, whiten=True, hid_size=[200, 100])

fig, ax = plt.subplots()
ax.plot(training_loss, label='training')
ax.plot(validation_loss, label='validation')
ax.set_xlabel('epoch')
ax.set_ylabel('loss')
ax.legend()
fig.tight_layout()
```
